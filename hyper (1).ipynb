```python
import pandas as pd

file_name = "تحليل هايبر.csv"

# استخدم المحرك "python" لتخمين الفاصل (delimiter)
df = pd.read_csv(file_name, sep=None, engine='python', encoding='utf-8-sig')

# عرض عينة من البيانات
display(df.head())
```

```python

df["سعر_الوحدة"] = pd.to_numeric(df["سعر_الوحدة"], errors="coerce")
df["عدد_الوحدات"] = pd.to_numeric(df["عدد_الوحدات"], errors="coerce")
df["الإيرادات"] = pd.to_numeric(df["الإيرادات"], errors="coerce")
df["التكلفة"] = pd.to_numeric(df["التكلفة"], errors="coerce")
df["الربح"] = pd.to_numeric(df["الربح"], errors="coerce")
```

```python

for c in df.select_dtypes(include=[float, int]).columns:
    if df[c].isna().any():
        df[c].fillna(df[c].median(), inplace=True)

print("✅ تم ملء القيم المفقودة في الأعمدة الرقمية بالوسيط.")
```

```python
display(df.head())
```

```python

for c in df.select_dtypes(include='object').columns:
    if df[c].isna().any():
        m = df[c].mode(dropna=True)
        if not m.empty:
            df[c].fillna(m.iloc[0], inplace=True)
```

```python
display(df.head())
```

```python
print(df.info())
```

```python
df["تاريخ_الشراء"] = pd.to_datetime(df["تاريخ_الشراء"], errors='coerce')
```

```python

for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.replace(r'\s+', ' ', regex=True).str.strip()
print(df.head())
```

```python

# 3 لحذف أي تكرارات بداخل الصفوف 
df = df.drop_duplicates()
display(df.head())
```
